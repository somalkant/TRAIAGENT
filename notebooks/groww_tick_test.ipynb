# Groww Tick Feed Test

Tests two approaches to getting live prices from Groww:

| | **Test A â€” REST polling** | **Test B â€” WebSocket (GrowwFeed)** |
|---|---|---|
| Method | `GrowwAPI.get_ltp()` every 1 s | `GrowwFeed.subscribe_ltp()` via NATS |
| Current code? | YES (current GrowwTickerAdapter) | NO (not implemented yet) |
| Expected latency | ~1-5 s (REST cache) | ~100 ms (real push feed) |

Run during market hours (9:15 AM â€“ 3:30 PM IST) for live data.

```python
import sys, os, json, time, threading
from pathlib import Path
from datetime import datetime

# Add project root to path
ROOT = Path(os.getcwd()).parent
if str(ROOT) not in sys.path:
    sys.path.insert(0, str(ROOT))

print(f"Project root: {ROOT}")
```

## 1. Load Groww access token

```python
from dotenv import load_dotenv
load_dotenv(ROOT / ".env")

from growwapi import GrowwAPI

# Try cached token first (written by live/agent.py --broker groww)
cache_path = ROOT / "checkpoints" / "access_token.json"
access_token = None

if cache_path.exists():
    cache = json.loads(cache_path.read_text())
    if cache.get("broker") == "groww":
        access_token = cache.get("access_token")
        print(f"Loaded token from cache (broker=groww)")

# Fall back to .env
if not access_token:
    access_token = os.getenv("GROWW_ACCESS_TOKEN", "")
    if access_token:
        print("Loaded token from .env GROWW_ACCESS_TOKEN")

if not access_token:
    raise RuntimeError(
        "No Groww token found. Run: python -m live.agent --broker groww  (then Ctrl+C after auth)"
    )

client = GrowwAPI(access_token)
print("GrowwAPI client created.")
```

## 2. Load exchange_tokens for test symbols

These come from the instruments.json cache (written by the live agent on startup).
If the cache is missing, we fetch fresh.

```python
# Symbols to test â€” a small focused set
TEST_SYMBOLS = ["KIRLOSENG", "RELIANCE", "TCS", "INFY", "HDFCBANK", "ICICIBANK", "SBIN"]

instruments_cache = ROOT / "checkpoints" / "instruments.json"

if instruments_cache.exists():
    data = json.loads(instruments_cache.read_text())
    if data.get("broker") == "groww":
        imap = data["map"]   # {symbol: exchange_token}
        print(f"Loaded {len(imap)} symbols from instruments.json (broker=groww)")
    else:
        print(f"Cache broker={data.get('broker')}, expected groww â€” fetching fresh")
        imap = None
else:
    print("No instruments.json cache â€” fetching fresh")
    imap = None

if imap is None:
    print("Fetching instrument list from Groww API...")
    from brokers.groww import GrowwClientAdapter
    adapter = GrowwClientAdapter()
    adapter.set_access_token(access_token)
    insts = adapter.instruments("NSE")
    imap = {i["tradingsymbol"]: i["instrument_token"] for i in insts}
    print(f"Fetched {len(imap)} NSE equity tokens")

# Build test instrument list
test_instruments = []
token_to_symbol = {}
groww_keys = []

for sym in TEST_SYMBOLS:
    token = imap.get(sym)
    if token:
        test_instruments.append({
            "exchange": "NSE",
            "segment": "CASH",
            "exchange_token": str(token),
        })
        token_to_symbol[str(token)] = sym
        groww_keys.append(f"NSE_{sym}")
        print(f"  {sym}: exchange_token={token}")
    else:
        print(f"  {sym}: NOT FOUND in instrument map")

print(f"\nReady to test {len(test_instruments)} symbols.")
```

## Test A â€” Current approach: REST polling via `GrowwAPI.get_ltp()`

This is exactly what `GrowwTickerAdapter._poll_loop()` does.
We poll 5 times with 1-second gaps and record timestamps.

```python
from brokers.groww import _parse_ltp_response

print("=== Test A: REST polling (current GrowwTickerAdapter approach) ===")
print(f"Symbols: {groww_keys}")
print()

poll_results = []

for poll_num in range(1, 6):
    t0 = time.time()
    try:
        resp = client.get_ltp(tuple(groww_keys), "CASH")
        elapsed = time.time() - t0
        now_str = datetime.now().strftime("%H:%M:%S.%f")[:-3]

        # Print raw response structure on first poll so we can see the format
        if poll_num == 1:
            sample_key = groww_keys[0]
            sample_val = resp.get(sample_key, "NOT FOUND")
            print(f"  [DEBUG] Response format for {sample_key}: type={type(sample_val).__name__}, value={sample_val}")
            print()

        poll_row = {"poll": poll_num, "clock": now_str, "elapsed_ms": round(elapsed * 1000)}
        for gk in groww_keys:
            d = resp.get(gk, 0)   # Groww returns float LTP directly, not a dict
            ltp, _ = _parse_ltp_response(d)
            sym = gk.replace("NSE_", "")
            poll_row[sym] = ltp

        poll_results.append(poll_row)

        prices = "  ".join(
            f"{gk.replace('NSE_', '')}={poll_row.get(gk.replace('NSE_', ''), 0):.2f}"
            for gk in groww_keys
        )
        print(f"  Poll {poll_num} [{now_str}] +{poll_row['elapsed_ms']}ms: {prices}")

    except Exception as e:
        import traceback
        print(f"  Poll {poll_num}: ERROR â€” {e}")
        traceback.print_exc()

    time.sleep(1.0)

print()
print("Test A done. Check if prices change between polls (should change if market is open).")
```

### Test A: Diagnose staleness

If prices are IDENTICAL across multiple polls during market hours â†’ Groww REST API is returning cached/delayed data.

```python
import pandas as pd

df_a = pd.DataFrame(poll_results).set_index("poll")

sym_cols = [s.replace("NSE_", "") for s in groww_keys]
print("REST polling results:")
print(df_a[["clock", "elapsed_ms"] + sym_cols].to_string())
print()

for sym in sym_cols:
    if sym in df_a.columns:
        prices = df_a[sym].tolist()
        unique = len(set(prices))
        verdict = "STALE (no change)" if unique == 1 else f"LIVE ({unique} unique values)"
        print(f"  {sym}: {prices} â†’ {verdict}")
```

## Test B â€” WebSocket approach: `GrowwFeed.subscribe_ltp()` via NATS

This uses Groww's real-time push feed (NATS over WSS). Not yet used in our live agent.
We subscribe, collect ticks for 30 seconds, then show results.

```python
import concurrent.futures
from growwapi.groww.feed import GrowwFeed

ws_ticks  = []
ws_lock   = threading.Lock()
groww_feed = None
OBSERVE_SECONDS = 20
CONNECT_TIMEOUT = 10   # seconds to wait for NATS connection

def on_tick_received(meta):
    now_str   = datetime.now().strftime("%H:%M:%S.%f")[:-3]
    token     = str(meta.get("feed_key", ""))
    sym       = token_to_symbol.get(token, token)
    try:
        all_ltp   = groww_feed.get_ltp()
        exchange  = meta.get("exchange", "NSE")
        segment   = meta.get("segment", "CASH")
        tick_data = all_ltp.get(exchange, {}).get(segment, {}).get(token, {})
        # Proto returns camelCase fields â€” try both
        ltp = float(
            tick_data.get("ltp") or tick_data.get("lastPrice") or
            tick_data.get("last_price") or 0
        )
        with ws_lock:
            ws_ticks.append({"clock": now_str, "symbol": sym, "ltp": ltp, "raw": tick_data})
        print(f"  WS [{now_str}] {sym}: {ltp:.2f}  raw_keys={list(tick_data.keys())[:8]}")
    except Exception as e:
        print(f"  WS [{now_str}] {sym}: callback error â€” {e}")


print("Creating GrowwFeed (NATS WebSocket)...")
print(f"  Connection timeout: {CONNECT_TIMEOUT}s")

def _create_feed():
    return GrowwFeed(client)

try:
    with concurrent.futures.ThreadPoolExecutor(max_workers=1) as ex:
        fut = ex.submit(_create_feed)
        try:
            groww_feed = fut.result(timeout=CONNECT_TIMEOUT)
        except concurrent.futures.TimeoutError:
            print(f"  TIMEOUT: NATS connection did not complete in {CONNECT_TIMEOUT}s")
            print("  Possible causes: Groww token expired, network issue, or server unreachable")
            ws_ticks = []
            raise RuntimeError("GrowwFeed connection timed out")

    print("GrowwFeed connected OK")

    print(f"Subscribing {len(test_instruments)} instruments...")
    result = groww_feed.subscribe_ltp(test_instruments, on_data_received=on_tick_received)
    print(f"Subscribe result: {result}")

    print(f"\nListening {OBSERVE_SECONDS}s for ticks (Ctrl+C / Interrupt to stop early)...\n")
    for i in range(OBSERVE_SECONDS):
        time.sleep(1)
        if (i + 1) % 5 == 0:
            print(f"  ... {i+1}s elapsed, {len(ws_ticks)} ticks so far")

    print(f"\nDone. Total ticks received: {len(ws_ticks)}")

except RuntimeError:
    pass   # already printed above
except Exception as e:
    print(f"GrowwFeed error: {type(e).__name__}: {e}")
    if "nacl" in str(e).lower() or "nkeys" in str(e).lower():
        print("  Fix: pip install pynacl nkeys")
    ws_ticks = []
```

### Test B: Results

```python
if ws_ticks:
    df_b = pd.DataFrame(ws_ticks)
    print(f"Total ticks received via WebSocket: {len(df_b)}")
    print()
    
    # Per-symbol summary
    print("Per-symbol tick count and price range:")
    for sym in TEST_SYMBOLS:
        sym_df = df_b[df_b["symbol"] == sym]
        if sym_df.empty:
            print(f"  {sym}: no ticks received")
        else:
            ltps  = sym_df["ltp"].tolist()
            first = sym_df.iloc[0]["clock"]
            last  = sym_df.iloc[-1]["clock"]
            print(f"  {sym}: {len(sym_df)} ticks  |  {first} â†’ {last}  |  "
                  f"min={min(ltps):.2f} max={max(ltps):.2f}")
    
    print()
    print("First 20 ticks:")
    print(df_b[["clock", "symbol", "ltp"]].head(20).to_string(index=False))
    
    print()
    print("Raw data fields from first tick (to understand proto structure):")
    print(df_b.iloc[0]["raw"])
else:
    print("No ticks received.")
    print("Possible reasons:")
    print("  1. Market is closed (test during 9:15 AM â€“ 3:30 PM IST)")
    print("  2. Token expired â€” re-run: python -m live.agent --broker groww")
    print("  3. Missing dependency: pip install pynacl nkeys")
```

## Comparison: REST polling vs WebSocket

```python
print("=" * 60)
print("COMPARISON SUMMARY")
print("=" * 60)
print()

# Test A summary
print("Test A â€” REST polling (current GrowwTickerAdapter):")
if poll_results:
    avg_ms = sum(r["elapsed_ms"] for r in poll_results) / len(poll_results)
    print(f"  API response time: ~{avg_ms:.0f} ms per call")
    stale_count = 0
    for sym in sym_cols:
        if sym in df_a.columns:
            prices = df_a[sym].tolist()
            if len(set(prices)) == 1:
                stale_count += 1
    if stale_count > 0:
        print(f"  {stale_count}/{len(sym_cols)} symbols had STALE prices (no change across 5 polls)")
    else:
        print(f"  All symbols showed price movement â€” REST data may be fresh")
else:
    print("  No results")

print()

# Test B summary
print("Test B â€” WebSocket / GrowwFeed (NATS push feed):")
if ws_ticks:
    df_b_summary = pd.DataFrame(ws_ticks)
    ticks_per_sym = df_b_summary.groupby("symbol").size()
    print(f"  Total ticks in {OBSERVE_SECONDS}s: {len(ws_ticks)}")
    print(f"  Avg ticks/symbol: {ticks_per_sym.mean():.1f}")
    print(f"  Ticks per symbol: {ticks_per_sym.to_dict()}")
    print(f"  â†’ WebSocket delivers real-time push (no polling delay)")
else:
    print("  No ticks received (market closed or auth issue)")

print()
print("CONCLUSION:")
print("  If REST prices are stale but WebSocket ticks are real-time,")
print("  upgrade GrowwTickerAdapter to use GrowwFeed.subscribe_ltp() instead.")
```
