# Phase 2 Paper Trades Analysis

Analyses `data/trade_logs/paper_trades.csv` â€” forward-only test results from 2023 onwards.  
Frozen weights from end of 2022. 1 trade per day. Strict no-lookahead.

**Results:** `EXACT_WIN` = target hit | `WIN` = profitable TIME_EXIT | `LOSS` = stopped / exit at loss  
**Conviction tiers:** `HIGH` (driver win%â‰¥65%) = 2Ã— risk | `MEDIUM` (â‰¥55%) = 1.5Ã— | `STANDARD` = 1Ã—

```python
import sys
from pathlib import Path

project_root = Path().resolve().parent
if str(project_root) not in sys.path:
    sys.path.insert(0, str(project_root))

import pandas as pd
import numpy as np
import plotly.graph_objects as go
from plotly.subplots import make_subplots

from config.settings import PAPER_TRADES_FILE

if not PAPER_TRADES_FILE.exists():
    print(f'Not found: {PAPER_TRADES_FILE}')
    print('Run: python run_testing.py 2023')
else:
    df = pd.read_csv(PAPER_TRADES_FILE, parse_dates=['date'])
    df['month']       = df['date'].dt.to_period('M').astype(str)
    df['year']        = df['date'].dt.year
    df['day_of_week'] = df['date'].dt.day_name()
    df['signal_dt']   = pd.to_datetime(df['signal_time'], format='%H:%M', errors='coerce')

    def _slot(t):
        if pd.isna(t): return 'unknown'
        h, m = t.hour, t.minute
        s = f'{h:02d}:{"00" if m < 30 else "30"}'
        eh, em = (h, 30) if m < 30 else (h+1, 0)
        return f'{s}-{eh:02d}:{em:02d}'
    df['time_slot'] = df['signal_dt'].apply(_slot)

    if 'conviction_tier' not in df.columns:
        df['conviction_tier'] = 'STANDARD'

    print(f'Loaded {len(df)} trades  |  {df["date"].min().date()} to {df["date"].max().date()}')
    df.head()
```

## 1. Overall Summary

```python
total    = len(df)
ew_count = (df['result'] == 'EXACT_WIN').sum()
w_count  = (df['result'] == 'WIN').sum()
l_count  = (df['result'] == 'LOSS').sum()
win_all  = ew_count + w_count
total_pnl = df['pnl_rs'].sum()
avg_win   = df.loc[df['pnl_rs'] > 0, 'pnl_rs'].mean()
avg_loss  = df.loc[df['pnl_rs'] < 0, 'pnl_rs'].mean()

print('=' * 55)
print('  PHASE 2 PAPER TRADE SUMMARY')
print('=' * 55)
print(f'  Total trades    : {total}')
print(f'  EXACT_WIN       : {ew_count:>4}  ({ew_count/total*100:.1f}%)  target hit')
print(f'  WIN (partial)   : {w_count:>4}  ({w_count/total*100:.1f}%)  profitable exit')
print(f'  LOSS            : {l_count:>4}  ({l_count/total*100:.1f}%)  stopped / exit at loss')
print(f'  Effective WR    : {win_all/total*100:.1f}%  (EXACT_WIN + WIN)')
print('-' * 55)
print(f'  Total P&L       : Rs {total_pnl:>10,.0f}')
print(f'  Avg win         : Rs {avg_win:>10,.0f}')
print(f'  Avg loss        : Rs {avg_loss:>10,.0f}')
print(f'  Best trade      : Rs {df["pnl_rs"].max():>10,.0f}')
print(f'  Worst trade     : Rs {df["pnl_rs"].min():>10,.0f}')
if avg_loss != 0:
    print(f'  Profit factor   : {abs(avg_win/avg_loss):.2f}')
print('=' * 55)
```

## 2. Cumulative P&L Over Time

```python
df_sorted = df.sort_values('date').copy()
df_sorted['cum_pnl'] = df_sorted['pnl_rs'].cumsum()
color_map = {'EXACT_WIN': 'lime', 'WIN': 'gold', 'LOSS': 'tomato'}

fig = go.Figure()
fig.add_trace(go.Scatter(x=df_sorted['date'], y=df_sorted['cum_pnl'],
    mode='lines', name='Cumulative P&L', line=dict(color='deepskyblue', width=2)))
for result, color in color_map.items():
    sub = df_sorted[df_sorted['result'] == result]
    fig.add_trace(go.Scatter(x=sub['date'], y=sub['cum_pnl'], mode='markers',
        name=result, marker=dict(color=color, size=7)))
fig.add_hline(y=0, line_dash='dash', line_color='gray')
fig.update_layout(title='Cumulative P&L', xaxis_title='Date', yaxis_title='Rs',
    template='plotly_dark', height=450, hovermode='x unified')
fig.show()
```

## 3. Result Breakdown

## 4. Monthly P&L

```python
monthly = df.groupby('month').agg(
    trades=('pnl_rs','count'), pnl=('pnl_rs','sum'),
    win_rate=('result', lambda x: (x != 'LOSS').mean() * 100)
).reset_index()
fig = make_subplots(rows=2, cols=1,
    subplot_titles=['Monthly P&L (Rs)', 'Monthly Win Rate (%)'],
    shared_xaxes=True, vertical_spacing=0.12)
fig.add_trace(go.Bar(x=monthly['month'], y=monthly['pnl'],
    marker_color=['lime' if v>=0 else 'tomato' for v in monthly['pnl']],
    text=[f'Rs {v:,.0f}' for v in monthly['pnl']], textposition='auto'), row=1, col=1)
fig.add_trace(go.Bar(x=monthly['month'], y=monthly['win_rate'],
    marker_color=['lime' if v>=50 else 'tomato' for v in monthly['win_rate']],
    text=[f'{v:.0f}%' for v in monthly['win_rate']], textposition='auto'), row=2, col=1)
fig.add_hline(y=50, row=2, col=1, line_dash='dash', line_color='gray')
fig.add_hline(y=0, row=1, col=1, line_dash='dash', line_color='gray')
fig.update_layout(template='plotly_dark', height=550, showlegend=False, title='Monthly Performance')
fig.show()
print(monthly.to_string(index=False))
```

## 5. Driver Strategy Performance

```python
strat_stats = df.groupby('driver_strategy').agg(
    count=('result','count'),
    exact_wins=('result', lambda x: (x=='EXACT_WIN').sum()),
    wins=('result', lambda x: (x=='WIN').sum()),
    losses=('result', lambda x: (x=='LOSS').sum()),
    total_pnl=('pnl_rs','sum'),
    avg_pnl=('pnl_rs','mean'),
    avg_rr=('rr','mean'),
).reset_index()
strat_stats['effective_wr']   = (strat_stats['exact_wins'] + strat_stats['wins']) / strat_stats['count'] * 100
strat_stats['exact_win_rate'] = strat_stats['exact_wins'] / strat_stats['count'] * 100
strat_stats = strat_stats.sort_values('count', ascending=False)

fig = make_subplots(rows=1, cols=2,
    subplot_titles=['Times Selected as Driver', 'Win Rate per Driver Strategy'])
fig.add_trace(go.Bar(y=strat_stats['driver_strategy'], x=strat_stats['count'],
    orientation='h', marker_color='deepskyblue'), row=1, col=1)
fig.add_trace(go.Bar(y=strat_stats['driver_strategy'], x=strat_stats['effective_wr'],
    orientation='h',
    marker_color=['lime' if v>=50 else 'tomato' for v in strat_stats['effective_wr']]), row=1, col=2)
fig.add_vline(x=50, row=1, col=2, line_dash='dash', line_color='gray')
fig.update_layout(template='plotly_dark', height=max(400, len(strat_stats)*30),
    title='Driver Strategy Analysis', showlegend=False)
fig.show()
print(strat_stats.sort_values('exact_win_rate', ascending=False)[
    ['driver_strategy','count','exact_wins','exact_win_rate','effective_wr','total_pnl','avg_rr']
].to_string(index=False, float_format='{:.1f}'.format))
```

## 14. Time of Day Analysis

Which 30-minute entry window has the highest win rate and average P&L?  
Opening volatility (9:15â€“9:45) vs confirmed trend (10:00+) â€” which works better?

```python
slot_order = [
    '09:15-09:45','09:45-10:15','10:15-10:45','10:45-11:15',
    '11:15-11:45','11:45-12:15','12:15-12:45','12:45-13:15',
    '13:15-13:45','13:45-14:15'
]
slot_stats = df.groupby('time_slot').agg(
    count=('result','count'),
    exact_wins=('result', lambda x: (x=='EXACT_WIN').sum()),
    win_rate=('result', lambda x: (x!='LOSS').mean()*100),
    exact_win_rate=('result', lambda x: (x=='EXACT_WIN').mean()*100),
    avg_pnl=('pnl_rs','mean'),
    total_pnl=('pnl_rs','sum'),
).reset_index()
slot_stats['slot_rank'] = slot_stats['time_slot'].apply(
    lambda s: slot_order.index(s) if s in slot_order else 99)
slot_stats = slot_stats.sort_values('slot_rank').reset_index(drop=True)

fig = make_subplots(rows=3, cols=1,
    subplot_titles=['Trades per Slot','Win Rate (%) per Slot','Avg P&L (Rs) per Slot'],
    shared_xaxes=True, vertical_spacing=0.08)
fig.add_trace(go.Bar(x=slot_stats['time_slot'], y=slot_stats['count'],
    marker_color='deepskyblue', text=slot_stats['count'], textposition='auto'), row=1, col=1)
fig.add_trace(go.Bar(x=slot_stats['time_slot'], y=slot_stats['win_rate'],
    marker_color=['lime' if v>=50 else 'tomato' for v in slot_stats['win_rate']],
    text=[f'{v:.0f}%' for v in slot_stats['win_rate']], textposition='auto'), row=2, col=1)
fig.add_trace(go.Bar(x=slot_stats['time_slot'], y=slot_stats['avg_pnl'],
    marker_color=['lime' if v>=0 else 'tomato' for v in slot_stats['avg_pnl']],
    text=[f'Rs {v:,.0f}' for v in slot_stats['avg_pnl']], textposition='auto'), row=3, col=1)
fig.add_hline(y=50, row=2, col=1, line_dash='dash', line_color='gray')
fig.add_hline(y=0, row=3, col=1, line_dash='dash', line_color='gray')
fig.update_layout(template='plotly_dark', height=650,
    title='Time of Day â€” Which Window is Best?', showlegend=False)
fig.show()

print('\nTime slot breakdown (sorted by win rate):')
print(slot_stats[['time_slot','count','exact_wins','exact_win_rate','win_rate','avg_pnl']]
    .sort_values('win_rate', ascending=False)
    .to_string(index=False, float_format='{:.1f}'.format))
```

## 15. Exact Target Hit Rate by Strategy

EXACT_WIN = price *actually reached* the predicted target. This is the purest signal quality measure â€”  
it shows whose **target formula** is most accurate, not just which strategy happened to be profitable.

```python
exact_stats = df.groupby('driver_strategy').agg(
    total=('result','count'),
    exact_wins=('result', lambda x: (x=='EXACT_WIN').sum()),
    total_pnl=('pnl_rs','sum'),
    avg_rr=('rr','mean'),
).reset_index()
exact_stats['exact_win_pct'] = exact_stats['exact_wins'] / exact_stats['total'] * 100
exact_stats = exact_stats[exact_stats['total'] >= 3].sort_values('exact_win_pct', ascending=False)

fig = make_subplots(rows=1, cols=2,
    subplot_titles=['EXACT_WIN Rate by Strategy (min 3 trades)','Total P&L by Strategy'])
fig.add_trace(go.Bar(
    y=exact_stats['driver_strategy'], x=exact_stats['exact_win_pct'], orientation='h',
    marker_color=['lime' if v>=50 else 'orange' if v>=35 else 'tomato' for v in exact_stats['exact_win_pct']],
    text=[f'{v:.0f}% ({int(n)} trades)' for v,n in zip(exact_stats['exact_win_pct'],exact_stats['total'])],
    textposition='auto'), row=1, col=1)
fig.add_trace(go.Bar(
    y=exact_stats['driver_strategy'], x=exact_stats['total_pnl'], orientation='h',
    marker_color=['lime' if v>=0 else 'tomato' for v in exact_stats['total_pnl']],
    text=[f'Rs {v:,.0f}' for v in exact_stats['total_pnl']], textposition='auto'), row=1, col=2)
fig.add_vline(x=50, row=1, col=1, line_dash='dash', line_color='gray')
fig.update_layout(template='plotly_dark', height=max(400, len(exact_stats)*35),
    title='Which Strategies Actually Hit Their Target?', showlegend=False)
fig.show()

print('\nTop 3 by EXACT_WIN rate:')
for _, row in exact_stats.head(3).iterrows():
    print(f'  {row["driver_strategy"]:<15}: {row["exact_win_pct"]:.0f}% exact  '
          f'({int(row["exact_wins"])}/{int(row["total"])} trades)  '
          f'Rs {row["total_pnl"]:,.0f} total P&L')
print('\nBottom 3 (targets rarely hit):')
for _, row in exact_stats.tail(3).iterrows():
    print(f'  {row["driver_strategy"]:<15}: {row["exact_win_pct"]:.0f}% exact  '
          f'({int(row["exact_wins"])}/{int(row["total"])} trades)')
```

## 16. Conviction Tier Performance

We risk more on HIGH-confidence drivers (VOL-SPIKE etc.).  
Does it actually pay off â€” do HIGH-tier trades win more often AND deliver better ROI?

```python
tier_order  = ['HIGH','MEDIUM','STANDARD']
tier_colors = {'HIGH':'lime','MEDIUM':'gold','STANDARD':'deepskyblue'}

tier_stats = df.groupby('conviction_tier').agg(
    count=('result','count'),
    exact_wins=('result', lambda x: (x=='EXACT_WIN').sum()),
    win_rate=('result', lambda x: (x!='LOSS').mean()*100),
    total_pnl=('pnl_rs','sum'),
    avg_pnl=('pnl_rs','mean'),
    avg_position=('position_rs','mean'),
).reindex(tier_order).reset_index()
tier_stats['exact_win_pct'] = tier_stats['exact_wins'] / tier_stats['count'] * 100
colors = [tier_colors.get(t,'gray') for t in tier_stats['conviction_tier']]

fig = make_subplots(rows=1, cols=3,
    subplot_titles=['Trade Count','Win Rate (%)','Avg P&L per Trade (Rs)'])
fig.add_trace(go.Bar(x=tier_stats['conviction_tier'], y=tier_stats['count'],
    marker_color=colors, text=tier_stats['count'], textposition='auto'), row=1, col=1)
fig.add_trace(go.Bar(x=tier_stats['conviction_tier'], y=tier_stats['win_rate'],
    marker_color=colors,
    text=[f'{v:.0f}%' for v in tier_stats['win_rate']], textposition='auto'), row=1, col=2)
fig.add_trace(go.Bar(x=tier_stats['conviction_tier'], y=tier_stats['avg_pnl'],
    marker_color=colors,
    text=[f'Rs {v:,.0f}' for v in tier_stats['avg_pnl']], textposition='auto'), row=1, col=3)
fig.add_hline(y=50, row=1, col=2, line_dash='dash', line_color='gray')
fig.add_hline(y=0, row=1, col=3, line_dash='dash', line_color='gray')
fig.update_layout(template='plotly_dark', height=420,
    title='Conviction Tier: Does Sizing Up Pay Off?', showlegend=False)
fig.show()

print('\nConviction tier summary:')
print(tier_stats[['conviction_tier','count','exact_win_pct','win_rate','avg_pnl','total_pnl','avg_position']]
    .to_string(index=False, float_format='{:.1f}'.format))
print('\nROI per tier (avg P&L / avg position):')
for _, row in tier_stats.iterrows():
    if row['avg_position'] > 0:
        roi = row['avg_pnl'] / row['avg_position'] * 100
        mult = {'HIGH':2.0,'MEDIUM':1.5,'STANDARD':1.0}.get(row['conviction_tier'],1.0)
        print(f'  {row["conviction_tier"]:<10}: {roi:.2f}% ROI  '
              f'(avg pos Rs {row["avg_position"]:,.0f} = {mult}x risk)')
```
